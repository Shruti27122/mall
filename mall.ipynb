```python
import pandas as pd
df=pd.read_csv('/home/pvg/sp/Mall_Customers.csv')
df
```

```python
df.isnull()
df
```

```python
df.isnull().sum()
```

```python
df['Genre'].replace({"Male":0,"Female":1},inplace=True)
df
```

```python
df1=df.iloc[:,3:5]
df1
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.scatterplot(data=df1,x='Annual Income (k$)',y='Spending Score (1-100)')
```

```python
from sklearn.cluster import KMeans
km=KMeans(n_clusters=3,init='k-means++',random_state=0).fit(df1)
km
```

```python
colors=km.labels_
centroids=km.cluster_centers_
x=[i[0] for i in centroids]
y=[i[1] for i in centroids]
```

```python
sns.scatterplot(data=df1,x='Annual Income (k$)',y='Spending Score (1-100)',hue=colors,palette=['violet','blue','orange'])
sns.scatterplot(x=x,y=y,marker="*",s=400,color='black')
```

```python
wcss=[] 
for i in range(1,11):
    km=KMeans(n_clusters=i,init='k-means++',random_state=0).fit(df1)
    wcss.append(km.inertia_)
wcss
```

```python
sns.lineplot(x=[i for i in range(1,11)],y=wcss).set(xlabel="Clusters",ylabel="WCSS")
```

```python
from sklearn.cluster import KMeans
km1=KMeans(n_clusters=5,init='k-means++',random_state=0).fit(df1)
km1
```

```python
colors1=km1.labels_
centroids1=km1.cluster_centers_
centroids1
```

```python
x1=[i[0] for i in centroids1]
y1=[i[1] for i in centroids1]
```

```python
sns.scatterplot(data=df1,x='Annual Income (k$)',y='Spending Score (1-100)',hue=colors1,palette=['violet','pink','orange','blue','grey'])
sns.scatterplot(x=x1,y=y1,marker='*',s=350,color='Black')
```

```python
import scipy.cluster.hierarchy as sh
sh.dendrogram(sh.linkage(df1,method='ward'))
```
